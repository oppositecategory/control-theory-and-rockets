```python
import numpy as np 
import matplotlib.pyplot as plt 


from scipy.integrate import odeint
```

```python
def double_integrator(x,t,u):
    q,dq = x 
    return [dq,u]

t = np.linspace(0,1,60)
x0 = [0,0.1]
u = 0
simulation = odeint(double_integrator,x0,t,args=(u,))
```

```python
x = np.linspace(-4,4,10)
y = np.linspace(-0.75,0.75,10)

xx,yy = np.meshgrid(x,y)

trajectories = np.zeros_like(xx)
u = 0
for i in range(10):
    for j in range(10):
        x0 = [xx[i,j],yy[i,j]]
        results = odeint(double_integrator,x0,[1,2],args=(u,))[-1]
        dx = results[0] - x0[0] + 1e-5
        dy = results[1] - x0[1] + 1e-5
        plt.arrow(xx[i,j],yy[i,j],dx,dy)
        
plt.grid()
plt.title(r"Phase portrait of double integrator around $u=0$")
plt.xlabel(r'$q$')
plt.ylabel(r'$\dot{q}$')
plt.show()
```
